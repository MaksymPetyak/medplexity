```python
# Parameters for evaluation
MODEL = "gpt-4"
DATASET = "mts-dialog"
SPLIT_TYPE = "test-1"
EVALUATION_FILE_PATH = f'{MODEL}-{DATASET}-{SPLIT_TYPE}.json'
```

```python
from medplexity.chains.summarisation_chain import SummarisationChain
from medplexity.llms.openai_caller import OpenAI
from medplexity.benchmarks.dataset_factory import DatasetFactory
from medplexity.medharness import Medharness

harness = Medharness(
    dataset=DatasetFactory().build(DATASET, SPLIT_TYPE),
    chain=SummarisationChain(
        llm=OpenAI(
            model=MODEL,
            api_token="",
            temperature=0,
        ),
        save_prompt=True
    ),
)
```

```python
harness.run(k=50, ignore_errors=True)
```

```python
harness.result
```

```python
from medplexity.prompts.conversation_summarisation_prompt_template import \
    ConversationSummarisationPromptTemplate
from datetime import datetime

harness.save_results(
    EVALUATION_FILE_PATH,
    additional_data={
        "prompt_template": ConversationSummarisationPromptTemplate.PROMPT,
        "date": datetime.now().date().strftime("%d-%m-%Y"),
    }
)
```
